Es un proceso mediante el cual se transforman los valores de un conjunto de datos para que todos estén en una misma escala o rango. Esto significa que, si tienes variables con diferentes magnitudes (por ejemplo, edad en años y salario en miles), la normalización ajusta esos valores para que sean comparables entre sí, generalmente llevándolos a un rango común como 
[
0
,
1
]
[0,1]. No cambia la información esencial de los datos, sino la forma en que están representados, facilitando su uso en análisis posteriores.

Sirve principalmente para mejorar el rendimiento de los algoritmos de análisis de datos y aprendizaje automático. Muchos modelos, especialmente los que dependen de distancias (como KNN o clustering), pueden verse afectados si una variable tiene valores mucho más grandes que otra, ya que dominaría el cálculo. Al normalizar, se evita este problema y se logra que todas las variables tengan el mismo peso en el análisis. Además, ayuda a que los modelos converjan más rápido y a obtener resultados más precisos y estables

```python
from sklearn import preprocessing
import numpy as np
```

```python
# Carga de datos.
X_train = np.array([[ 1., -4.,  5.],
                    [ 5.,  0.,  2.],
                    [ 0.,  6., -3.]])

print(X_train)
```

```python
# Limpieza de datos: normalización.
normalizer = preprocessing.MinMaxScaler()
X_train_norm = normalizer.fit_transform(X_train)
```

```python
# Salida de resultados.
print(X_train_norm)
```

```python
# Ejemplo para ver que los resultados son iguales si aplicamos los métodos "fit" y "transform" por separado
X_train_2 = np.copy(X_train)

# Crea un objeto "fitted_normalizer" llamando al método "fit"
f = normalizer.fit(X_train)

# Aplica el objeto sobre los datos "X_train_2" para obtener "X_train_norm_2"
X_train_norm_2 = f.transform(X_train)

# Comprueba si "X_train_norm" y "X_train_norm_2" son iguales
print(np.array_equiv(X_train_norm, X_train_norm_2))
```
