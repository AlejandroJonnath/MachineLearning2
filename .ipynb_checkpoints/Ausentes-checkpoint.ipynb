#### ¿Qué son los Datos Ausentes?

Son aquellos valores que no están presentes en el conjunto de datos donde deberían existir, es decir, es información que falta en una o más variables de tu dataset.

#### ¿Cómo se ven los datos ausentes?

- NaN (Not a Number)
- None
- Celdas vacías
- Valores especiales como -999 (en algunos datasets)

#### Ejemplo

| Edad | Ingresos | Ciudad    |
| ---- | -------- | --------- |
| 25   | 500      | Quito     |
| NaN  | 700      | Guayaquil |
| 30   | NaN      | Quito     |

Aquí faltan datos en:

Edad (fila 2)
Ingresos (fila 3)

#### ¿Por qué ocurren?

- Errores al recolectar datos
- Personas que no responden encuestas
- Fallos en sensores o sistemas
- Datos no aplicables (ej: “embarazo” en hombres)

```python
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
# Carga de datos
X_train = [[1,      4], 
           [np.nan, 3], 
           [7,      8]]
```

```python
# Limpieza de datos: imputación valores ausentes (modelo)
imputer = SimpleImputer(strategy='mean') # Utilizamos la estrategia de la media

# Aplicar los métodos "fit" y "transform" para imputar los valores pérdidos de X_train
X_train_clean = imputer.fit_transform(X_train)
print(X_train_clean)
```

```python
# Imputación de valores ausentes sobre el conjunto de test
X_test = [[np.nan, 10], 
          [6,      np.nan], 
          [8,      2]]

# Imputar los datos del conjunto de test
X_test_clean=imputer.transform(X_test)
print(X_test_clean)
```
